#### Cleaning

```python
# Count duplicates based on all columns
num_duplicates = df.duplicated().sum()
print(f"Number of duplicates based on all columns: {num_duplicates}")
```

#### Prime genre and user rating summary

These findings can give insights into user preferences and can help developers to prioritize certain categories when designing and releasing new apps

```python
# Create a pivot table that summarizes user_rating by prime_genre
genre_rating_pivot = df.pivot_table(values='user_rating', index='prime_genre')

# Sort the pivot table by mean user_rating in descending order
genre_rating_pivot_sorted = genre_rating_pivot.sort_values('user_rating', ascending=False)

# Print the results
print(genre_rating_pivot_sorted)
```

#### Distribution of app ratings

These shall help to see how the app ratings are distributed. This can give an idea of whether the ratings are mostly positive or negative, and if there are any outliers.

```python
import seaborn as sns

# Create a distribution plot of app ratings
sns.histplot(df['user_rating'], kde=False)

# Add plot labels and title
plt.xlabel('User Rating')
plt.ylabel('Number of Apps')
plt.title('Distribution of App Ratings')

# Show the plot
plt.show()
```

#### Price distribution by category

This would help us to identify categories with the highest and lowest app prices, and also show whether the price distribution is skewed or has outliers

```python
import seaborn as sns

# Create a violin plot of app prices grouped by category
sns.violinplot(x='price', y='prime_genre', data=df)

# Set x-axis label and y-axis label
plt.xlabel('Price')
plt.ylabel('Category')

# Add a title to the plot
plt.title('Price Distribution by Category')

# Show the plot
plt.show()
```

#### Correlation analysis

This would help us to see how different variables are related to each other.

```python
# Calculate the correlation matrix
corr = df.corr()

# Create a heatmap to visualize the correlation matrix
fig, ax = plt.subplots(figsize=(12, 8), dpi=100)
sns.heatmap(corr, cmap='coolwarm', annot=True, ax=ax)
plt.xticks(rotation=45)

# Add a title to the plot
plt.title('Correlation Matrix')

# Show the plot
plt.show()
```

#### Top app categories

This can help to identify popular app categories and target those categories for new app development.

```python
# Count the number of apps in each category
category_counts = df['prime_genre'].value_counts()

# Sort the results in descending order
category_counts = category_counts.sort_values(ascending=False)

# Print the top 10 categories
print(category_counts.head(10))
```

#### App size by category

This would help to identify categories with the largest and smallest app sizes

```python
# Group the data by category and calculate the average app size for each category
category_sizes = df.groupby('prime_genre')['size_bytes'].mean()

# Convert the app sizes from bytes to megabytes
category_sizes = category_sizes / (1024 * 1024)

# Sort the results in descending order
category_sizes = category_sizes.sort_values(ascending=False)

# Print the results
print(category_sizes)
```
